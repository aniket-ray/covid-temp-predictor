```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
import math
import datetime
```

```python
df = pd.read_csv('./dataset/imu_ToBeSplit.csv')
```

```python
df['phase'] = df['phase'].astype(int)
```

```python
dataset = df.copy()
dataset.drop('time', axis = 1, inplace=True)
dataset
```

```python
def window_len_mean(df):
    first = df['phase'][0]
    count = 0
    N_STRIDES = 0
    mean = 0

    for phase in df['phase']:
        if (phase == first):
            count += 1
        else:
            mean += count
            N_STRIDES += 1
            first = phase
            count = 0
    mean += count
    mean = (mean)/(N_STRIDES + 1)

    return mean
```

```python
WINDOW_SIZE = 60

N_FEATURES = len(dataset.columns) - 1
STEP = 1

segments = []
labels = []

for i in range(0, len(dataset) - WINDOW_SIZE, STEP):
    feature = [] #For every window size
    for column in dataset:
        if column == 'phase':
            break
        feature.append(dataset[column].values[i : i + WINDOW_SIZE])
    label = stats.mode(dataset['phase'][i : i + WINDOW_SIZE])
    label = label[0][0] 
    segments.append(feature)
    labels.append(label)

segments = np.array(segments)
labels = np.array(labels)
```

```python
segments.shape
segments = np.asarray(segments, dtype = np.float32).reshape(-1, WINDOW_SIZE, N_FEATURES)
```

```python
print(segments[0][0])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(segments, labels, test_size = 0.2, shuffle = False)
```

```python
X_train.shape
```

```python
X_test, Y_test = shuffle(X_test, Y_test)
```

```python
X_train = tf.keras.utils.normalize(X_train, axis = 1)
X_test = tf.keras.utils.normalize(X_test, axis = 1)
```

```python
%load_ext tensorboard
!rm -rf ./logs/

verbose, epochs, batch_size = 1, 30, 128

WINDOW_SIZE =  X_train.shape[1]
N_FEATURES = X_train.shape[2]
N_OUTPUTS = 9

model = tf.keras.models.Sequential()

model.add(tf.keras.layers.LSTM(100, input_shape=(WINDOW_SIZE, N_FEATURES)))

model.add(tf.keras.layers.Dense(100, activation='relu'))
#model.add(tf.keras.layers.Dense(100, activation='relu'))
#model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(N_OUTPUTS, activation='softmax'))

adam = tf.keras.optimizers.Adam(learning_rate=0.005, beta_1=0.9, beta_2=0.999, amsgrad=False)
model.compile(loss='sparse_categorical_crossentropy',
              optimizer= adam,
              metrics=['accuracy'])


log_dir="logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
    # fit network
model.fit(X_train, Y_train,
          epochs=epochs, 
          batch_size = batch_size, 
          verbose=verbose, 
          callbacks=[tensorboard_callback],
          validation_data=(X_test, Y_test))
```

```python
I = 10
predct = model.predict(X_test[I].reshape(1,60,18))
print(np.argmax(predct))

print(Y_test[I])
```

```python
score, accuracy = model.evaluate(X_test, Y_test, batch_size=batch_size, verbose=1)
print(score)
print(accuracy)
model.save('gaitPrediction.h5') 
```
